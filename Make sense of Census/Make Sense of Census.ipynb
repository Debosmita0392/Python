```python
# Importing header files
import numpy as np
import warnings

warnings.filterwarnings('ignore')

#New record
new_record=[[50,  9,  4,  1,  0,  0, 40,  0]]

path = 'C:\\Users\\rajro\\Documents\\Glabs\\Basecamp-master\\Make sense of Census\\file.csv'

#Reading file
data = np.genfromtxt(path, delimiter=",", skip_header=1)
#print(data)
#Code starts here
new_record=np.array(new_record)
#print(new_record)
census = np.concatenate([data,new_record]) #appending one new row to existing array
#print(census.shape)
age = census[:,0]
print(age)
max_age = np.max(age)
print(max_age)
min_age = np.min(age)
print(min_age)
age_mean = np.mean(age)
print(round(age_mean,2))
age_std = np.std(age)
print(round(age_std,2))
race_0=census[census[:,2]== 0]
len_0 = len(race_0)
print(len_0)
race_1=census[census[:,2] == 1]
len_1 = len(race_1)
print(len_1)
race_2=census[census[:,2] == 2]
len_2 = len(race_2)
print(len_2)
race_3=census[census[:,2] == 3]
len_3 = len(race_3)
print(len_3)
race_4=census[census[:,2] == 4]
len_4 = len(race_4)
print(len_4)
a = np.array([len_0,len_1,len_2,len_3,len_4])
print(a)
if len_0 == a.min():
    minority_race = 0
elif len_1 == a.min():
    minority_race = 1
elif len_2 == a.min():
    minority_race = 2
elif len_3 == a.min():
    minority_race = 3
else:
    minority_race = 4
print(minority_race)
senior_citizens = census[census[:,0] > 60]
print(senior_citizens)
working_hours_sum = np.sum(senior_citizens[:,6])
print(working_hours_sum)
senior_citizens_len = len(senior_citizens)
print(senior_citizens_len)
avg_working_hours = working_hours_sum/senior_citizens_len
print(round(avg_working_hours,2))
if avg_working_hours > 25 :
    print("Govt policy is not followed")
else :
    print("Govt policy is followed")
high = census[census[:,1] > 10]
low = census[census[:,1] <= 10]
avg_pay_high = np.mean(high[:,7])
print(round(avg_pay_high,2))
avg_pay_low = np.mean(low[:,7])
print(round(avg_pay_low,2))
```
