```python
# Importing header files
import numpy as np
import warnings

warnings.filterwarnings('ignore')

#New record
new_record=[[50,  9,  4,  1,  0,  0, 40,  0]]

path = 'C:\\Users\\rajro\\Documents\\Glabs\\Basecamp-master\\Make sense of Census\\file.csv'

#Reading file and load dataset in a variable
data = np.genfromtxt(path, delimiter=",", skip_header=1) # header is skipped because it contains string values and numpy can store only data of single datatype
#print(data)
#Code starts here
new_record=np.array(new_record)
#print(new_record)
census = np.concatenate([data,new_record]) #appending one new row to existing array
#print(census.shape)
#We often associate the potential of a country based on the age distribution of the people residing there. 
#Do a simple analysis of the age distribution
age = census[:,0]
print(age) # Array containiing only age data
max_age = np.max(age)
print("MAximum Age is ", max_age)
min_age = np.min(age)
print("Minimum Age is ",min_age)
age_mean = np.mean(age)
print("Mean Age is ",round(age_mean,2))
age_std = np.std(age)
print("Standard Deviation is ",round(age_std,2))
#The constitution of the country tries it's best to ensure that people of all races are able to live harmoniously
#Check the country's race distribution to identify the minorities so that the government can help them
race_0=census[census[:,2]== 0]
len_0 = len(race_0)
#print(len_0)
race_1=census[census[:,2] == 1]
len_1 = len(race_1)
#print(len_1)
race_2=census[census[:,2] == 2]
len_2 = len(race_2)
#print(len_2)
race_3=census[census[:,2] == 3]
len_3 = len(race_3)
#print(len_3)
race_4=census[census[:,2] == 4]
len_4 = len(race_4)
#print(len_4)
a = np.array([len_0,len_1,len_2,len_3,len_4])
#print(a)
if len_0 == a.min():
    minority_race = ' Amer-Indian-Eskimo'
elif len_1 == a.min():
    minority_race = 'Asian-Pac-Islander'
elif len_2 == a.min():
    minority_race = 'Black'
elif len_3 == a.min():
    minority_race = 'Other'
else:
    minority_race = 'White'
print("Minority race is ",minority_race)
#As per the new govt. policy, all citizens above age 60 should not be made to work more than 25 hours per week.
#Analyse the data and see if that policy is followed.
senior_citizens = census[census[:,0] > 60]
#print(senior_citizens)
working_hours_sum = np.sum(senior_citizens[:,6])
#print(working_hours_sum)
senior_citizens_len = len(senior_citizens)
#print(senior_citizens_len)
avg_working_hours = working_hours_sum/senior_citizens_len
print("Average working hours is ",round(avg_working_hours,2))
if avg_working_hours > 25 :
    print("Hence Govt policy is not followed")
else :
    print("Hence Govt policy is followed")
#Our parents have repeatedly told us that we need to study well in order to get a good(read: higher-paying) job.
#Analyse whether the higher educated people have better pay in general.
high = census[census[:,1] > 10]
low = census[census[:,1] <= 10]
avg_pay_high = np.mean(high[:,7])
print("Average pay of highly educated people ", round(avg_pay_high,2))
avg_pay_low = np.mean(low[:,7])
print("Average pay of less educated people ",round(avg_pay_low,2))
```
